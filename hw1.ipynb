```python

import pandas as pd
import openpyxl
from os import rename
import numpy as np
%matplotlib qt5
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
dfdata = pd.read_excel('D:\Pycharm by URFU\Задание по очистке данных.xlsx', na_values=['-'])   #'-' = Nan
dfdata.info()   #в 0-ой колонке все данные равны NaN
```

```python
dfdata
```

```python
clean_df = dfdata.copy()
clean_df = clean_df.drop(dfdata.columns[0], axis=1)   #удаление 0-ой колонки с NaN значениями
clean_df
```

```python
clean_df.dropna(axis=0, inplace=True)   #удаление всех NaN по строкам
clean_df

```

```python
#создание колонок с годом, месяцем и днём на основании колонки с датой, затем удаление колонки с датой
clean_df['Year'] = clean_df['Date'].dt.year
clean_df['Month'] = clean_df['Date'].dt.month
clean_df['Day'] = clean_df['Date'].dt.day
clean_df = clean_df.drop(['Date'], axis=1)
```

```python
clean_df.info()
```

```python
#создание колонок с часом и минутами на основании колонки со временем, удаление колонки со временем
time_df = clean_df['Time'].str.split(':', expand=True)
clean_df['Hour'] = time_df[0].astype(int)
clean_df['Minute'] = time_df[1].astype(int)
clean_df = clean_df.drop(['Time'], axis=1)
```

```python
clean_df
```

```python
#оставляем только м/с в колонке со скоростью
clean_df['Wind_speed_2'] = clean_df['Wind_speed_2'].str.split(' / ', expand=True)[1]
```

```python
clean_df
```

```python
#снова делим колонку со скоростьюпо пробелу, зменяем возможные пустые значения на Nan для их последующего удаления, приводим к float
clean_df['Wind_speed_2'] = clean_df['Wind_speed_2'].str.split(' ', expand=True)[0].replace('', np.nan).astype(float)
```

```python
#колонка со скоростью имеет 4 Nan, которые нужно удалить
clean_df['Wind_speed_2'].isna().sum()
```

```python
#удаляем все получившиеся Nan по строкам
clean_df.dropna(axis=0, inplace=True)
```

```python
clean_df.info()
```

```python
#изменяем название колонки со скоростью
clean_df.rename(columns={'Wind_speed_2': 'Wind_speed_m/s'}, inplace=True)
```

```python
clean_df
```

```python
#из колонки с номером дня уберём символ d разделением строки по нему, возьмём только правую часть[1]
clean_df['Day_number'] = clean_df['Day_number'].str.split('d', expand=True)[1].astype(int)
```

```python
#из колонок с данными широты и долготы оставляем только число также разделением строки по пробелу,оставляем правую часть[1]
clean_df['Latitude'] = clean_df['Latitude'].str.split(' ', expand=True)[1].str.replace(',', '.').astype(float)
```

```python
clean_df
```

```python

clean_df['Longitude'] = clean_df['Longitude'].str.split(' ', expand=True)[1].str.replace(',', '.').astype(float)
```

```python
clean_df
```

```python
#записываем получившийся результат в файл без индексов
clean_df.to_excel('clean_data.xlsx', index=False)
```
